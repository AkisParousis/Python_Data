```python
#Major with highest mid-career salary.
print(df['Undergraduate Major'].loc[df['Mid-Career Median Salary'].idxmax()])
print(df['Mid-Career Median Salary'].loc[df['Mid-Career Median Salary'].idxmax()])
```

```python
pd.reset_option('^display.', silent=True)
#A low-risk major is a degree where there is a small difference between the lowest and highest salaries.
spread_col = df['Mid-Career 90th Percentile Salary'].subtract(df['Mid-Career 10th Percentile Salary'])
df.insert(1, 'Spread', spread_col)
df.head()
```

```python
#Sort with ascending order, to see which degrees have the smallest spread
sort_df = df.sort_values('Spread', ascending=True)
sort_df[['Undergraduate Major', 'Spread']].head()
```

```python
#Sort df to see majors with the highest potential
sort_df = df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
sort_df[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#Group data to see how many majors we have in each group
df.groupby('Group').count()
```

```python
pd.options.display.float_format = '{:,.2f}'.format 
df.groupby('Group').mean()
```
